```python
!pip install pandas scikit-learn

import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python

data = {
    'Movie A': [5, 4, 0, 0],
    'Movie B': [4, 0, 0, 2],
    'Movie C': [0, 0, 5, 4],
    'Movie D': [0, 3, 4, 5]
}
```

```python
df = pd.DataFrame(data,
                  index=['User1', 'User2', 'User3', 'User4'])

print("Dataset:\n")
print(df)
```

```python
similarity = cosine_similarity(df.T)
```

```python
similarity_df = pd.DataFrame(
    similarity,
    index=df.columns,
    columns=df.columns
)

print("\nMovie Similarity Matrix:\n")
print(similarity_df)
```

```python
movie_name = 'Movie A'

print(f"\nRecommended movies for {movie_name}:\n")

print(
    similarity_df[movie_name]
    .sort_values(ascending=False)
)
```
